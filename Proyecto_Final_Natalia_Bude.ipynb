5.3. Estableciendo el contexto y el prompt. Solicitud de texto a texto con openAI.

```python
#Estableciendo el contexto de nuestro entorno inmediato
context = 'Actúa como un asesor financiero profesional, experto en transmitir conocimientos sobre finanzas personales.'

#prompt (text/user's question)
prompt = "Transmite de forma clara, concisa sobre finanzas personales y enséñanos el ABC, el paso a paso, el camino para convertir a una persona que gestiona muy bien sus propias finanzas. Además, indícanos de forma clara, esquemática y generalizada de como deberíamos de gestionar nuestros gastos personales de manera eficiente. El contenido no puede supera los 3400 caracteres."
```

```python
# Make the request
conversation = [
    {"role": "system", "content": context},
    {"role": "user", "content": prompt}
]

# Solicitud para obtener la respuesta
response = openai.ChatCompletion.create(
    model='gpt-3.5-turbo',
    messages=conversation,
    max_tokens=1500
)

message = response.choices[0]['message']
print("{}: {}".format(message['role'], message['content']))
```

5.4. Estableciendo contexto y prompt para generar una imagen.

```python
#Make the context of our prompt envoirement
context = 'Ahora actúa como un diseñador y dibujante experto.'

#prompt (text/user's question)
prompt_image = "Crea una imagen que transmita una atmósfera cálida y acogedora, en la que personas de diferentes edades y perfiles estén gestionando sus finanzas personales de manera segura y relajada. La escena debe mostrar a personas estableciendo metas, revisando presupuestos y ahorrando, con una actitud positiva y confiada sobre su futuro. Utiliza una paleta de colores suaves y armónicos para generar un sentido de tranquilidad y optimismo. Incluye elementos visuales simbólicos que representen el crecimiento financiero, como plantas que florecen, gráficos ascendentes o monedas creciendo en una hucha. La imagen debe evocar hábitos financieros saludables y generar una conexión emocional que inspire confianza en la planificación financiera."
```

5.5. Settings Generación de la imágen con OpenAI

```python
# Make the request for the image prompt
image_response = openai.Image.create(
    prompt=prompt_image,
    n=1,
    size="512x512"
)

# Get and print the generated image URL
print(image_response['data'][0]['url'])
```

5.6. Solicitud para descargar la imagen.

```python
# Download the logo
response = requests.get(image_response['data'][0]['url'])

# Verify if possible to download
if response.status_code == 200:
    # Directory to save the image
    directory = "./generated_images/"

    # Get filename
    filename = os.path.basename(image_response['data'][0]['url'])
        # Agregar la extensión ".png"
    filename = "imagen.png"
       
    # Make directory if not exists
    if not os.path.exists(directory):
        os.makedirs(directory)
    
    # Complete path
    filepath = os.path.join(directory, filename)
    
    # Save the image
    with open(filepath, 'wb') as f:
        f.write(response.content)
        
    print("La imagen ha sido descargada y almacenada correctamente en:", filepath)
    img = Image.open(filepath)
    print(f"Prompt utilizado: {prompt_image}")
    display(img)

else:
    print("Hubo un error al descargar la imagen.")
```
